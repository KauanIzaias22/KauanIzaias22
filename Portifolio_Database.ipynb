```python
import streamlit as st
from pymongo import MongoClient

# MongoDB connection
# The connection string points to the 'mongodb' service defined in docker-compose.yml
# and uses the default MongoDB port 27017.
try:
    client = MongoClient('mongodb://mongodb:27017/')
    db = client['mydatabase']  # Replace 'mydatabase' with your desired database name
    collection = db['mycollection'] # Replace 'mycollection' with your desired collection name
    st.success("Connected to MongoDB!")
except Exception as e:
    st.error(f"Could not connect to MongoDB: {e}")

# CRUD Functions

def create_document(data):
    """Creates a new document in the MongoDB collection."""
    try:
        result = collection.insert_one(data)
        return result.inserted_id
    except Exception as e:
        st.error(f"Error creating document: {e}")
        return None

def read_documents():
    """Reads all documents from the MongoDB collection."""
    try:
        documents = list(collection.find())
        return documents
    except Exception as e:
        st.error(f"Error reading documents: {e}")
        return []

def update_document(document_id, new_data):
    """Updates a document in the MongoDB collection."""
    try:
        # MongoDB stores _id as ObjectId, so we need to convert the input document_id
        from bson.objectid import ObjectId
        result = collection.update_one({'_id': ObjectId(document_id)}, {'$set': new_data})
        return result.modified_count
    except Exception as e:
        st.error(f"Error updating document: {e}")
        return 0

def delete_document(document_id):
    """Deletes a document from the MongoDB collection."""
    try:
        # MongoDB stores _id as ObjectId, so we need to convert the input document_id
        from bson.objectid import ObjectId
        result = collection.delete_one({'_id': ObjectId(document_id)})
        return result.deleted_count
    except Exception as e:
        st.error(f"Error deleting document: {e}")
        return 0

# Streamlit UI (Basic example, will be expanded later)
st.title("MongoDB CRUD App")

st.header("Create Document")
# Example form for creating a document
with st.form("create_form"):
    key = st.text_input("Key")
    value = st.text_input("Value")
    submitted = st.form_submit_button("Create")
    if submitted:
        if key and value:
            new_doc = {key: value}
            inserted_id = create_document(new_doc)
            if inserted_id:
                st.success(f"Document created with ID: {inserted_id}")
        else:
            st.warning("Please enter both key and value.")

st.header("Read Documents")
if st.button("Refresh Documents"):
    documents = read_documents()
    if documents:
        st.write("Documents in collection:")
        for doc in documents:
            st.write(doc)
    else:
        st.info("No documents found.")

st.header("Update Document")
# Example form for updating a document
with st.form("update_form"):
    update_id = st.text_input("Document ID to Update")
    update_key = st.text_input("Key to Update")
    update_value = st.text_input("New Value")
    submitted_update = st.form_submit_button("Update")
    if submitted_update:
        if update_id and update_key and update_value:
            updated_data = {update_key: update_value}
            modified_count = update_document(update_id, updated_data)
            if modified_count > 0:
                st.success(f"Document with ID {update_id} updated.")
            else:
                st.warning(f"Document with ID {update_id} not found or no changes made.")
        else:
            st.warning("Please enter Document ID, Key, and New Value.")

st.header("Delete Document")
# Example form for deleting a document
with st.form("delete_form"):
    delete_id = st.text_input("Document ID to Delete")
    submitted_delete = st.form_submit_button("Delete")
    if submitted_delete:
        if delete_id:
            deleted_count = delete_document(delete_id)
            if deleted_count > 0:
                st.success(f"Document with ID {delete_id} deleted.")
            else:
                st.warning(f"Document with ID {delete_id} not found.")
        else:
            st.warning("Please enter Document ID to Delete.")
```

## Integração da aplicação streamlit com o docker

### Subtask:
Criar um `Dockerfile` para a aplicação Streamlit, definindo as dependências e o ponto de entrada da aplicação, e atualizar o `docker-compose.yml` para incluir o serviço da aplicação Streamlit, garantindo que ele se conecte ao contêiner do MongoDB.


**Reasoning**:
Create the Dockerfile for the Streamlit application and update the docker-compose.yml file to include the Streamlit service, linking it to the MongoDB service.



```python
# Create the Dockerfile content
dockerfile_content = """
FROM python:3.9-slim

WORKDIR /app

COPY requirements.txt .

RUN pip install --no-cache-dir -r requirements.txt

COPY app.py .

EXPOSE 8501

CMD ["streamlit", "run", "app.py", "--server.port", "8501", "--server.address", "0.0.0.0"]
"""

# Write the content to a file named Dockerfile
with open('Dockerfile', 'w') as f:
    f.write(dockerfile_content)

print("Created Dockerfile for Streamlit application.")

# Create a requirements.txt file (assuming streamlit and pymongo are the dependencies)
requirements_content = """
streamlit
pymongo
"""
with open('requirements.txt', 'w') as f:
    f.write(requirements_content)

print("Created requirements.txt file.")


# Read the existing docker-compose.yml content
with open('docker-compose.yml', 'r') as f:
    docker_compose_content = f.read()

# Add the streamlit service to the docker-compose.yml content
# We need to carefully add this after the 'services:' key and before 'volumes:'
streamlit_service_content = """
  streamlit_app:
    build: .
    ports:
      - "8501:8501"
    depends_on:
      - mongodb
"""

# Find the position to insert the new service
insert_position = docker_compose_content.find('volumes:')

if insert_position != -1:
    # Insert the streamlit service content before the 'volumes:' section
    updated_docker_compose_content = (
        docker_compose_content[:insert_position] +
        streamlit_service_content +
        docker_compose_content[insert_position:]
    )
else:
    # If 'volumes:' is not found, just append the service content
    updated_docker_compose_content = docker_compose_content + streamlit_service_content


# Write the updated content back to docker-compose.yml
with open('docker-compose.yml', 'w') as f:
    f.write(updated_docker_compose_content)

print("Updated docker-compose.yml with Streamlit service.")
```

**Reasoning**:
Execute the docker-compose build and up commands to build the images and start the containers.



```python
import subprocess

# Build the Docker images
print("Building Docker images...")
build_process = subprocess.run(['docker-compose', 'build'], capture_output=True, text=True)
print(build_process.stdout)
if build_process.stderr:
    print(build_process.stderr)

if build_process.returncode == 0:
    print("Docker images built successfully. Starting services...")
    # Start the services in detached mode
    up_process = subprocess.run(['docker-compose', 'up', '-d'], capture_output=True, text=True)
    print(up_process.stdout)
    if up_process.stderr:
        print(up_process.stderr)

    if up_process.returncode == 0:
        print("Docker services started successfully. Verify with 'docker ps'.")
    else:
        print("Failed to start Docker services.")
else:
    print("Failed to build Docker images.")
```
